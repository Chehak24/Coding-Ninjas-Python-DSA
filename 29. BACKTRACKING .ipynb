### MAZE N x N
1. print all paths, either go left,right,yop or bottom
2. 0 is a blocking path
3. if path is chosen, put all other ele as zero so taht only 1s are taken to be printed, 
4. explore all the paths from every 4 direction
5. stop at n-1 x n-1


```python
def printPathHelper(x,y,maze,n,solution):
    if x==n-1 and y ==n-1: ## rchd the destination
        solution[x][y]=1
        print(solution)
        return
    if x<0 or y<0 or x>=n or y>=n or maze[x][y]==0 or solution[x][y]==1:## original maze has blocking pts
     ## this pt is already being included
        return
    
    ##blocked all thre points
    solution[x][y]=1 ## initialising mmyself since I have rchd the desired pt
    printPathHelper(x+1,y,maze,n,solution)
    printPathHelper(x,y+1,maze,n,solution)
    printPathHelper(x-1,y,maze,n,solution)
    printPathHelper(x,y-1,maze,n,solution)
    
    solution[x][y]=0 ## coz backtracking krke jab wapis jaa re ho after exploring all the paths, then jis path se aaye the vo zero ho jaana chaie
    return
def printPath(maze):
    n=len(maze)
    
    ## solution matrix has all zeroes, not initialising with anything
    
    solution=[[0 for j in range(n)]for i in range(n)]
    
    printPathHelper(0,0,maze,n,solution)

n= int(input())
maze=[]
for i in range(n):
    row =[int(ele) for ele in input().split()]
    maze.append(row)
    
printPath(maze)
```

### N QUEENS
1. N X N chessboard -> such that no queen cross each other
2. Eg: suppose a 4x4 chessboard, put queens in places of horse directions to each other
3. Q can move in all possible 8 directions

```python
def isSafe(row,col,board,n):
    ## vertical direction
    i=row-1
    while i>=0:
        if board[i][col]==1:
            return False
        i-=1
    
    ## upper diagonal left
    i=row-1
    j=col-1
    while i>=0 and j>=0:
        if board[i][j]==1:
            return False
        
        i-=1
        j-=1
    
    ## upper diagonal right
    i=row-1
    j=col+1
    while i>=0 and j<n:
        if board[i][j]==1:
            return False
        i-=1
        j+=1
    return True

def printPathsHelper(row,n,board):
    if row==n:
        for i in range(n):
            for j in range(n):
                print(board[i][j],end=' ')
        print()
        return
    
    for col in range(n):
        if isSafe(row,col,board,n):
            board[row][col]=1
            printPathsHelper(row+1,n,board)
            
            board[row][col]=0 ## backtracking requires it to put it to 0
            ##taaki wapis jab usss row pr aaye vo 0 hi le usse
    return
        

def printPaths(n):
    board=[[0 for j in range(n)]for i in range (n)]
    printPathsHelper(0,n,board)

n = int(input())
printPaths(n)
```
