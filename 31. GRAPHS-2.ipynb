#### BUILDING MST FROM graph
1. INPUT:
1. 1--->2 (wt 6)
    1. 2-->3 (wt 2)
    2. 1-->3 (wt 2)
    3. 1-->0  (wt 2)

2. OUTPUT:
    1. 2-->3 (wt 2)
    2. 1-->3 (wt 2)
    3. 1-->0 (wt 2)

1. 1-->2 (wt 6 is not included)

```python
class Edge:
    def __init__(self,src,dest,wt):
        self.src=src
        self.dest=dest
        self.wt=wt
        
 
li=[int(ele) for ele in input().split()]
n=li[0]
E=li[1]
edges=[]

def getParent(v,parent):
    if v==parent[v]: ## if the v actual parent is it itself, that is the main parent
        return v  
    return getParent(parent[v],parent)

def krushkal(edges,nVertices):
    parent=[i for i in range(nVertices)]
    edges=sorted(edges,key=lambda edge:edge.wt) ## 
    count=0 ### no.of elements that will be included in the MST
    output=[]
    
    i=0 ## iterating on edges
    while count<(nVertices-1):
        currentEdge=edges[i]
        srcParent=getParent(currentEdge.src,parent)
        destParent=getParent(currentEdge.dest,parent)
        
        if srcParent!=destParent: ## if parents r diff
            output.append(currentEdge) 
            count+=1 ## since output is appended 
            parent[srcParent]=destParent ## src parent badlega to dest parent mei
            
        i+=1
    return output

for i in range(E):
    curr_input=[int(ele) for ele in input().split()]
    src=curr_input[0]
    dest=curr_input[1]
    wt=curr_input[2]
    
    edge = Edge(src,dest,wt)
    edges.append(edge)
    
output=kruskal(edges,n)

for edge in output: ## output format
    if edge.src<edge.dest:
        print(str(edge.src)+' '+str(edge.dest)+' '+str(edge.wt))
    else:
        print(str(edge.dest)+' '+  str(edge.src)+' '+str(edge.wt))
        
```
