```python
## if any compression, merging ie a change has happened,
## ie a movement has happened, then only a new 2 will be added
import random
def start_game():
    mat=[]
    for i in range(4):
        mat.append([0]*4)
    return mat

def add_new2(mat):
    ##generating random space
    r=random.randint(0,3)
    c=random.randint(0,3)
    
    ## inserting only at places with an empty space
    while mat[r][c]!=0:
        r=random.randint(0,3)
        c=random.randint(0,3)
    mat[r][c]=2
    
    ## current state
def get_current_state(mat):
    for i in range(4):
        for j in range(4):
            if mat[i][j]==2048:
                return 'Won'
    ## if any place 0 is present ie there r empty positions
    for i in range(4):
        for j in range(4):
            if mat[i][j]==0:
                return 'GAME NOT OVER'
            
    ## if adjacent ele same, that means those numbers can still be combined
    for i in range(3): ## 3 so that matrix doesnt go outside
        for j in range(3):
            if(mat[i][j]==mat[i+1][j] or mat[i][j]==mat[i][j+1]):
                return 'GAME NOT OVER'
    ## but last row can have adjacent ele as  well
    for j in range(3):
         if mat[3][j]==mat[3][j+1]:
                return 'GAME NOT OVER'
    ## Last Column
    for i in range(3):
        if mat[i][3]==mat[i+1][3]:
            return 'GAME NOT OVER'
        
    return 'Lost'

## COMPRESSING IE NON ZEROES NUMBERR SHOULD BE ON ONE SIDE

def compress(mat):
    changed=False
    new_mat=[]
    for i in range(4):
        new_mat.append([0]*4)
        
    for i in range(4):
        pos=0
        for j in range(4):
            if mat[i][j]!=0:
                new_mat[i][pos]=mat[i][j]
                if j!=pos: ## compression has takenn place 
                    changed=True
                pos+=1
    return new_mat,changed

def reverse(mat):
    new_mat=[]
    for i in range(4):
        new_mat.append([])
        for j in range (4):
            new_mat[i].append(mat[i][4-j-1]) ## reversing the elements in same row
    return new_mat

def transpose(mat):
    new_mat=[]
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[j][i])
            
    return new_mat
## ELEMENTS COMBINING
def merge(mat):
    changed=False
    for i in range(4):
        for j in range(3):
            if mat[i][j]==mat[i][j+1] and mat[i][j]!=0:
                mat[i][j]=mat[i][j]*2
                mat[i][j+1]=0
                changed=True
    return mat,changed

def move_up(grid):
    transpose_grid=transpose(grid)
    new_grid,changed1=compress(transpose_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid=compress(new_grid)
    final_grid=transpose(new_grid)
    return final_grid,changed

def move_down(grid):
    transpose_grid=transpose(grid)
    reversed_grid=reverse(transpose_grid)
    new_grid,changed1=compress(reversed_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid=compress(new_grid)
    final_reverse_grid=reverse(new_grid)
    final_grid=transpose(final_reverse_grid)
    return final_grid,changed

def move_right(grid):
    reversed_grid=reverse(grid)
    new_grid,changed1=compress(reversed_grid)
    new_grid,changed2=merge(new_grid)
    changed=changed1 or changed2
    new_grid=compress(new_grid)
    final_grid=reverse(new_grid)
    return final_grid,changed

def move_left(grid):
    new_grid,changed1=compress(grid)
    new_grid,chnaged2=merge(new_grid)
    changed=changed1 or changed2 ## ie any compression or merging has happened
    new_grid,temp=compress(new_grid) ## coz this will happen inly when merging has happened
    return new_grid,changed
```
