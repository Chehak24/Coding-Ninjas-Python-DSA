### HAS PATH ?
1. To return True or False
2. if v1 and v2 are adjacent ie edge, there is a path
3. if not adjacent, v1 checking if there is a path to v2 thru other vertices ie adjacent vertices of v1 have path to v2, follow dfs 
4. hasPath(v1,v2) :
            1. v1 and v2 are adjacent : return True
            2. for all adjacent vertices of v1 which aint visited: hasPath(a,v2)? True: return True
            return False
            

```python
import queue

class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjMatrix=[[0 for i in range(nVertices+1)] for j in range(nVertices+1)]
        
    def addEdge(self,v1,v2):
        self.adjMatrix[v1][v2]=1
        self.adjMatrix[v2][v1]=1
        
    def __hasPathHelper(self,v1,v2,visited):
        if self.adjMatrix[v1][v2]==1:
            return True
        
        else:
            visited[v1]=True
            for i in range(self.nVertices):
                if self.adjMatrix[v1][i]>0 and visited[i] is False:
                    flag=self.__hasPathHelper(i,v2,visited)
                    if flag:
                        return flag
        return False
    
    
    def hasPath(self,v1,v2):
        visited=[False for i in range(self.nVertices)]
        return self.__hasPathHelper(v1,v2,visited)
    
v=int(input())
e=int(input())

g=Graph(v)

for i in range(e):
    li=list(input().split())
    x=int(li[0])
    y=int(li[1])
    g.addEdge(x,y)
v1=int(input())
v2=int(input())

ans=g.hasPath(v1,v2)

if ans:
    print('true')
else:
    print('false')
```

### GET PATH IN DFS
1. last ele can have multiple paths
2. we hav to get the path following the order that vertex with lesser value is called first
3. path's 2 condns : v1 and v2 are adjacent, adjacent vertices of v1 have path to v2, if adjacent isnt the case, return is None, next vertex to call
4. 2 has a path to 6 (0->2->4->6) 2 will return [6,4,2] to 0 which is not None, and hence 0 itself will get appended ie [6,4,2,0]

1. getPath(v1,v2):
        1. if v1 and v2 are adjacent: return [v2 v1]
        2. explore all adjacent path: 
                1. (not visited) 
                2. either return None or if u get a path, append urself and return; no need to go further
                

```python
import queue

class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjMatrix=[[0 for i in range(nVertices)] for j in range(nVertices)]
        
    def addEdge(self,v1,v2):
        self.adjMatrix[v1][v2]=1
        self.adjMatrix[v2][v1]=1
        
    def __getPathHelper(self,v1,v2,visited):
        if v1==v2:
            path=[]
            path.append(v2)
            return path
        visited[v1]=True
        for i in range(self.nVertices):
            if self.adjMatrix[v1][i]>0 and visited[i] is False:
                path=self.__getPathHelper(i,v2,visited)
                if path is not None:
                    path.append(v1)
                    return path
        return None
    
    def getPath(self,v1,v2):
        visited=[False for i in range(self.nVertices+1)]
        return self.__getPathHelper(v1,v2,visited)
    
    def __str__(self):
        return str(self.adjMatrix)
    
v=int(input())
e=int(input())

g=Graph(v)

for i in range(e):
    li=list(input().split())
    x=int(li[0])
    y=int(li[1])
    g.addEdge(x,y)
v1=int(input())
v2=int(input())

ans=g.getPath(v1,v2)

if ans is not None:
    for i in ans:
        print(i,end=' ')
```

```python
class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjMatrix=[[0 for i in range(nVertices+1)] for j in range(nVertices+1)]
        
    def addEdge(self,v1,v2):
        self.adjMatrix[v1][v2]=1
        self.adjMatrix[v2][v1]=1
        
    def __isConnected(self,v,visited):
        visited[v]=True
        for i in range(self.nVertices):
            if self.adjMatrix[v][i]>0 and visited[i] is False:
                self.__isConnected(i,visited)
                
        
    def isConnected(self):
        visited=[False for i in range(self.nVertices)]
        if len(visited)==0:
            return True
        self.__isConnected(0,visited)
        for i in visited:
            if not i:
                return False
        return True
    
v=int(input())
e=int(input())

g=Graph(v)

for i in range(e):
    li=list(input().split())
    x=int(li[0])
    y=int(li[1])
    g.addEdge(x,y)
    
if g.isConnected():
    print('true')
    
else:
    print('false')

```
