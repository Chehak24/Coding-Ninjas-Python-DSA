```python
##reverse linked list O(N^2)
class Node:
    def __init__ (self,data):
        self.data=data
        self.next=None
        
def printll(head):
    while head is not None:
        print(head.data,end=" ")
        head=head.next
    return head

def takeinput():
    inputList=[int(ele) for ele in input().split()]
    head = None
    tail = None
    for currdata in inputList:
        if currdata==-1:
            break
        newNode=Node(currdata)
        if head is None:
            head=newNode
        else:
            curr = head
            while curr.next is not None:
                curr=curr.next
            curr.next = newNode
    return head
def reverse (head):
    if head is None or head.next is None:
        return head
    smallhead=reverse(head.next)
    curr=smallhead
    while curr.next is not None:
        curr=curr.next ##high complexity coz of this step
    curr.next=head
    head.next = None
    return smallhead

head=takeinput()   
printll(head)
head=reverse(head)
printll(head)
```
