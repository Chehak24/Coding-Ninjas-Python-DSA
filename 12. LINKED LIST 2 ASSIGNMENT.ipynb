```python
# KREVERSE

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def takeinput():
    inputlist=[int(ele) for ele in input().split()]
    head=None
    tail=None
    for currdata in inputlist:
        if currdata== -1:
            break
        newnode=Node(currdata)
        if head is None:
            head=newnode
            tail=newnode
        else:
            tail.next=newnode
            tail=newnode
    return head
def printll(head):
    while head is not None:
        print(head.data,end=" ")
        head=head.next
    print()
    
def reverse(head):
    if head is None or head.next is None:
        return head,head
    smallhead,smalltail=reverse(head.next)
    smalltail.next=head
    head.next=None
    return smallhead,head

def kreverse(head,k):
    if head is None or head.next is None:
        return head
    count=1
    head1= head
    tail1=head
    while tail1.next is not None:
        if count==k:
            break
        count+=1
        tail1=tail1.next
    head2=tail1.next
    tail1.next=None
    newhead,newtail=reverse(head1)
    newtail.next=kreverse(head2,k)
    return newhead

head=takeinput()
printll(head)
printll(kreverse(head,3))
```

```python
##swap nodes
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def printll(head):
    while head is not None:
        print(head.data,end=" ")
        head=head.next
    print()
def takeinput():
    inputlist=[int(ele) for ele in input().split()]
    head=None
    tail=None
    for currdata in inputlist:
        if currdata==-1:
            break
        newnode=Node(currdata)
        if head is None:
            head=newnode
            tail=newnode
        else:
            tail.next=newnode
            tail=newnode
    return head
def swapNodes(head,i,j):
    if i == j:
        return head
    curr=head
    prev=None
    firstnode=None
    secondnode=None
    firstnodeprev=None
    secondnodeprev=None
    pos=0
    while curr is not None:
        if pos==i:
            firstnodeprev=prev
            firstnode=curr
        elif pos==j:
            secondnodeprev=prev
            secondnode=curr
        prev=curr
        curr=curr.next
        pos+=1
    if firstnodeprev is not None:
        firstnodeprev.next=secondnode
    else:
        head=secondnode
    if secondnodeprev is not None:
        secondnodeprev.next=firstnode
    else:
        head=firstnode
    currfirstnode=secondnode.next
    secondnode.next=firstnode.next
    firstnode.next=currfirstnode
    return head
head=takeinput()
printll(head)
i=int(input())
j=int(input())
newhead=swapNodes(head,i,j)
printll(newhead)
```
